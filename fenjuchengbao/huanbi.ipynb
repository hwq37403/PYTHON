```python
#合并数据
def merge_excel(file,outpath):

    li=pd.DataFrame()
    for i in file:
        df_tmp=pd.read_excel(i,sheet_name='服务中心汇总',skiprows=0,header=[0,1,2])
        li=pd.concat([li,df_tmp])

    writer = pd.ExcelWriter(r'E:\WORK\网格酬金核算\test/'+outpath)
    li.to_excel(writer,'Sheet1')
     
    writer.save()
    writer.close()
```
