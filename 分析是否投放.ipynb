```python
def flagjudge(test):
    fmt='%Y-%m-%d %H:%M:%S'
    fmt_time='{0}-{1}-{2} {3}:{4}:{5}'
    d = time.strftime(fmt, time.localtime())
    d2=time.strptime(d, fmt)
    d1=time.strptime(test, fmt)#需要判断的时间

    if d2.tm_mon-1==0:
         m=12
    else:
        m=d2.tm_mon-1
    lastMonthDay=calendar.monthrange(d2.tm_year,m)[1]

    d_time=fmt_time.format(d2.tm_year,m,lastMonthDay,23,59,59)
    d_time=time.strptime(d_time, fmt)#上月最后一天
    print(d1)
    print(d_time)
    if d1>d_time:
        return '是'
    else:
        return '否'
```

```python
a='2017-12-31 00:00:00'
print(flagjudge(a))
```
