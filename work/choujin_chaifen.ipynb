```python
import time  # 引入time模块

import pandas as pd

import numpy as np

import os

import openpyxl

import shutil

from openpyxl import load_workbook

import tkinter as tk
from tkinter import filedialog

###由于PD会覆盖重写sheet，为了保存一个EXCEL多个sheet，所有只能通过EXCELwriter使用一个句柄来写入多个sheet

'''
区县分局承包拆分表格，拆分时需对四个子表的区县字段进行升序排序，保证顺序一致，否则每个区县的子表顺序不一致。
'''
```

```python
l=path.rfind('/')

tmp_path=path[0:l+1]+'模板.xlsx'
```

```python

file1 = pd.read_excel(path, sheet_name='6-渠道酬金汇总报表')
file_t=file1.copy()

file_qx=file_t[file_t['区县名称'].notnull()]
file_fwzx=file_t[file_t['服务中心'].notnull()]


file_qx.columns=file_qx.columns.str.strip()
file_fwzx.columns=file_fwzx.columns.str.strip()
```

```python
file_qx=file_qx.sort_values(by='区县名称')
file_fwzx=file_fwzx.sort_values(by='服务中心')
```

```python
menu1 = file_qx.loc[:,'区县名称'].drop_duplicates()


```

```python
menu1
```

```python
# 去掉合计
menu1=menu1[~ menu1.str.contains('合计') ]
```

```python
menu1
```

```python
#创建到区县目录
path=path[0:l]+"/拆分"
if os.path.exists(path)==False:
    os.mkdir(path)

for name1 in menu1:
    tpath=path+'/'+name1
    if os.path.exists(tpath)==False:
        os.mkdir(tpath)
```

```python
menu2 = file_fwzx.loc[:, ['区县名称','服务中心','渠道名称']].drop_duplicates()
for index, row in menu2.iterrows():
        # print(menu2)
        # input()
        # print (row["服务中心"], row["渠道名称"])
        
        #处理僰字乱码
        if row["服务中心"].find('?')>=0:
            row["服务中心"]=row["服务中心"].replace('?','僰')
        if row["渠道名称"].find('?')>=0:
            row["渠道名称"]=row["渠道名称"].replace('?','僰')

        #TODO 自助创建模板
        filename=tmp_path
        qx_path=path[0:l]+"/拆分/"+row["区县名称"]
        fwzx_path=qx_path+"/"+row["服务中心"]
        outpath = fwzx_path+"/"+ row["渠道名称"] + ".xlsx"
        #复制模板到拆分
        # print(outpath)
        
        #创建到服务中心目录
        if os.path.exists(outpath)==False:
            if os.path.exists(fwzx_path)==False:
                os.mkdir(fwzx_path)

        shutil.copyfile(filename, outpath)

        # 读取被写入的Excel工作簿
        book = load_workbook(outpath)
        # 建立写入对象
        write = pd.ExcelWriter(outpath, engine='openpyxl')
        write.book = book
        write.sheets = {ws.title: ws for ws in book.worksheets}


        
        df_res = file_fwzx[file_fwzx['渠道名称'] == row['渠道名称']]
        

        df_res.to_excel(write, sheet_name='6-渠道酬金汇总报表', header=False, index=False, startrow=3, startcol=0)
        
        
        write.save()
        if count<row_len:
            count+=1
            print('拆分进度',count)

    



print('拆分完成')

```
